```python
import statistics
madrs_avg=[]
for i in range(0,len(df)):
  data=[df['madrs1'][i],df['madrs2'][i]]
  #print(data)
  madrs_avg.append(statistics.mean(data))
df['MADRS']=madrs_avg
df
```

```python

df=df.drop("madrs1",axis=1)
df=df.drop('madrs2',axis=1)

#df=df.drop('index',axis=1)
df=df.drop('days',axis=1)

#df=df.drop('gender',axis=1)
#df=df.drop("afftype",axis=1)
#df=df.drop("melanch",axis=1)
#df=df.drop("inpatient",axis=1)
```

```python
import statistics
Age=[]
for i in df['age']:
  x=[int(i[0:2]),int(i[3:])]
  Age.append(statistics.mean(x))
df['Age']=Age
df=df.drop('age',axis=1)
df
```

```python
Edu=[]
for i in df['edu']:
  if(len(i)==4):
    x=int(i[0:1])
    Edu.append(x)
  else:
    x=int(i[0:2])
    Edu.append(x)

df['Edu']=Edu
df=df.drop('edu',axis=1)
df
```

```python
df.corr()
```

```python
resting=[24,22,23,20,19,23,24,23,18,18,30,23,26,31,26,60,22,24,21]
df['Resting_period']=resting
df.head()
```
